```python
import sys
import pennylane as qml
from pennylane import numpy as np
from QMLatomLoader import *
from tqdm import tqdm
from scipy.optimize import minimize

n_qubit=3

init_params=np.abs(np.random.random(n_qubit*2+1,requires_grad=True))
opt = qml.AdagradOptimizer(stepsize=0.5)
params=init_params
batch=64

def cost(para):
    param=np.abs(para)
    return paraoptim(batchs=batch,classic_parameter=param[1:].reshape((n_qubit,2)), weigthed=True,cutoff_radius=param[0])
```

```python
epochs=1000
losslist=[]
for i in tqdm(range(epochs)):
    if i%10==0:
        
        params,loss=opt.step_and_cost(cost,params)
        print(loss)
        losslist.append(loss)
        p=np.array(losslist)
        print(p)
        print(params)
    
    else:
        params=opt.step(cost,params)
        params=np.abs(params)
```
